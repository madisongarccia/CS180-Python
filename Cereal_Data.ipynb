## Exercise 4: America
Quantify the relationship between sugar and ratings. 

Make a plot to visualize this relationship. Superimpose a best fit line (with seaborn) to describe the relationship. It may be helpful to look at the [seaborn regplot documentation]("https://seaborn.pydata.org/generated/seaborn.regplot.html").

```python
# Make the plot for the data visualization and line of best fit here:
sns.regplot(data = cereal, x = 'sugars', y = 'rating')
```


Calculate a correlation statistic describing the relationship between sugar and ratings  (i.e., r or r squared).

```python
# Calculate the statistic using this cell:
cereal['sugars'] = cereal['sugars'].astype(float)
cereal['rating'] = cereal['rating'].astype(float)
np.corrcoef(cereal['sugars'], cereal['rating'])[0][1]
```

Write a statement in plain English interpreting this statistic.

The correlation coefficient is -0.76 which means there is a strong negative relationship between sugar content and ratings, meaning as sugar increases, ratings tend to decrease for that brand. 

## Exercise 5: America Part 2
Make five plots comparing the relationships of carbo, sugars, calories, protein, and fat with rating.

```python
# Write your code to compare the various variables with rating below:
fig, ax = plt.subplots(nrows=3, ncols=2, figsize=(15, 15))

ax[0][0].scatter(cereal[['sugars']], cereal[['rating']] )
ax[0][0].set_ylabel("rating")
ax[0][0].set_xlabel("sugars")
ax[0][0].title.set_text("relationship between sugar and rating")

ax[0][1].scatter(cereal[['carbo']], cereal[['rating']] )
ax[0][1].set_ylabel("rating")
ax[0][1].set_xlabel("carbo")
ax[0][1].title.set_text("relationship between carbs and rating")

ax[1][1].scatter(cereal[['calories']], cereal[['rating']] )
ax[1][1].set_ylabel("rating")
ax[1][1].set_xlabel('calories')
ax[1][1].title.set_text("relationship between calories and rating")

ax[1][0].scatter(cereal[['protein']], cereal[['rating']] )
ax[1][0].set_ylabel("rating")
ax[1][0].set_xlabel('protein')
ax[1][0].title.set_text("relationship between protein and rating")

ax[2][0].scatter(cereal[['fat']], cereal[['rating']] )
ax[2][0].set_ylabel("rating")
ax[2][0].set_xlabel('fat')
ax[2][0].title.set_text("relationship between fat and rating")
```

Of the variables carbo, sugars, calories, protein, and fat, which has the strongest relationship with rating? Justify your answer.

sugar and rating seem to have the strongest relationship with a strong negative linear relationship. This indicates that the higher the sugar, the lower the ratings tend to be for that brand. 

## Exercise 6: Preparing for Mt. Everest
Do the same as you did with exercise 5, but instead compare carbo, sugars, protein, and fat with calories.

```python
# Write your code to compare the various variables with calories below:
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(15, 15))

ax[0][0].scatter(cereal[['sugars']], cereal[['calories']] )
ax[0][0].set_ylabel("calories")
ax[0][0].set_xlabel("sugars")
ax[0][0].title.set_text("relationship between sugar and calories")

ax[0][1].scatter(cereal[['carbo']], cereal[['calories']] )
ax[0][1].set_ylabel("calories")
ax[0][1].set_xlabel("carbo")
ax[0][1].title.set_text("relationship between carbs and calories")

ax[1][1].scatter(cereal[['fat']], cereal[['calories']] )
ax[1][1].set_ylabel("calories")
ax[1][1].set_xlabel('fat')
ax[1][1].title.set_text("relationship between fat and calories")

ax[1][0].scatter(cereal[['protein']], cereal[['calories']] )
ax[1][0].set_ylabel("calories")
ax[1][0].set_xlabel('protein')
ax[1][0].title.set_text("relationship between protein and calories")
```

Of the variables carbo, sugars, protein, and fat, which has the strongest relationship with calories? Justify your answer.

sugars seems to have the strongest positive correlation to calories, because the scatterplot shows the most distinct behavior. This would also make sense that as a cereal's sugar content increases, the calories also increases.
